Proses pra-pemrosesan data yang dilakukan dalam penelitian ini mencakup beberapa tahap, seperti pengolahan emoji; mengubah huruf menjadi huruf kecil; menghapus nama pengguna, angka, dan tanda baca; memperbaiki kesalahan pengejaan dan singkatan, serta menghilangkan spasi berlebih; menghapus kata-kata pengisi (stopwords), dan mengakhiri kata (stemming).

```python
empos = re.compile(r"(😂)|(❤)|(♥)|(😍)|(😘)|(😊)|(👌)|(💕)|(👏)|(😁)")
emneg = re.compile(r"(😭)|(😩)|(😒)|(😔)|(😡)|(😴)|(🔫)|(😞)|(😪)|(😫)")
emnet = re.compile(r"(☯)|(✨)|(★)|(█)|(🔥)|(♫)|(�)|(©)|(👀)|(🐓)")

def emoticon(txt):
    a = ""
    while a !=txt:
        a = empos.sub(" positif", txt)
        txt = a
        a = empos.sub(" positif", a)
        b = ""
        while b !=txt:
            b = emneg.sub(" negatif", txt)
            txt = b
            b = emneg.sub(" negatif", b)
            c = ""
            while c !=txt:
                c = emnet.sub(" netral", txt)
                txt = c
                c = emnet.sub(" netral", c)
            return txt

mark = re.compile(r"[0-9]|[^\w\s]|_")
mark1 = re.compile(r"'")

def rep(txt):
    a = ""
    while a !=txt:
        a = mark.sub(" ", txt)
        txt = a
        a = mark.sub(" ", a)
        b = ""
        while b !=txt:
            b = mark1.sub(" ", txt)
            txt = b
            b = mark1.sub(" ", b)
            c = ""
        return txt

def load(filename):
    f = open(filename,"r")
    dic = {}
    for line in f:
        a,b = line.split(" = ")
        dic[a]= b[:-1]
    f.close()
    return dic

# Membaca file teks
with open('abbreviations_rev2.txt', 'r') as file:
    lines = file.readlines()

# Inisialisasi kamus kosong
abfile = {}

# Mengolah setiap baris dalam file
for line in lines:
    # Memisahkan kata dan artinya menggunakan tanda sama dengan (=)
    parts = line.strip().split('=')
    
    # Pastikan ada dua bagian (kata dan arti) dalam setiap baris
    if len(parts) == 2:
        word = parts[0].strip()
        meaning = parts[1].strip()
        
        # Menambahkan pasangan kata dan arti ke dalam kamus
        abfile[word] = meaning

def change(text):
    word = text.split()
    l = len(word)
    for i in range(l):
        if word[i] in abfile:
            word[i]=abfile[word[i]]
    return " ".join(word)

def hilang_repetisi(text):
    kata=text.split()
    l=len(kata)
    for i in range(l):
        S=kata[i]
        ll=len(S)
        j=0
        k=1
        R=S[0]
        while k<ll:
            if S[j]!=S[k]:
                R+=S[k]
                j=k
            k+=1
        kata[i]=R
    return " ".join(kata)


def load(filename):
	f=open(filename,"r")
	ret=set()
	for line in f:
		ret.add(hilang_repetisi(line[:-1]))
	f.close
	return ret

stoplist= pd.read_csv("stopwords_uci.txt")

def rem_stopword(txt):
	words=txt.split()
	l=len(words)
	for i in range(l):
		if words[i] in stoplist:
			words[i]=""
	return " ".join(words)

# create stemmer
stem = StemmerFactory()
stemmer = stem.create_stemmer()

def stemInd(text):
    words = text.split()
    words = ([stemmer.stem(w) for w in words])
    return " ".join(words)

stemm = SnowballStemmer('english')
def stemEng(text):
    words = text.split()
    words = ([stemm.stem(w) for w in words])
    return " ".join(words)

```
